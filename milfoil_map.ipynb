```python
import pandas as pd
import numpy as np
import folium

df = pd.read_csv('milfoil_tableB_all_years_cleaned.csv')
df
```

```python
lake_map_2008 = folium.Map(location=[42.950253, -75.870825], zoom_start=13)
lake_map_2008
```

```python
df['milfoil_presence'].unique()
```

```python
df['milfoil_presence'] = df['milfoil_presence'].fillna('N')
df
```

```python
df2008 = df.loc[df.year == 2008]
```

```python
# none, trace, sparse, moderate, dense
milfoil_abundance = ['N', 'T', 'S', 'M', 'D'] # use index for number value

# for ma_index, value in enumerate(milfoil_abundance):
#   for df_index, presence in enumerate(df.milfoil_presence):
#     if presence == value:
#       print(ma_index, df['latitude'][df_index], df['longitude'][df_index])
```

```python
for ma_index, value in enumerate(milfoil_abundance):
  for df_index, presence in enumerate(df2008.milfoil_presence):
    if presence == value:
      folium.features.CircleMarker(
        [df2008['latitude'][df_index], df2008['longitude'][df_index]],
        radius=ma_index*2,
        color=None, #'green', 
        fill_color='green', 
        fill_opacity=0.5,
    ).add_to(lake_map_2008)
lake_map_2008
```

```python
df2009 = df.loc[df.year == 2009]
lake_map_2009 = folium.Map(location=[42.950253, -75.870825], zoom_start=13)
# lake2009
```

```python
for ma_index, value in enumerate(milfoil_abundance):
  for df_index, presence in enumerate(df2009.milfoil_presence):
    # print(type(df2009['latitude']))
    if presence == value:
      folium.features.CircleMarker(
        [df2009['latitude'].item(), df2009['longitude']].item(),
        radius=ma_index*2,
        color=None,
        fill_color='green', 
        fill_opacity=0.5,
    ).add_to(lake_map_2009)
lake_map_2009
```
