```python
import pandas as pd
import numpy as np
import folium

df = pd.read_csv('milfoil_tableB_all_years_cleaned.csv')
df
```

```python
df['corrected_longitude'] = np.where(df['longitude'] > 0, df['longitude']*-1, df['longitude'])
df
```

```python
lake_map_2008 = folium.Map(location=[42.950253, -75.870825], zoom_start=13)
lake_map_2008
```

```python
df['milfoil_presence'].unique()
```

```python
df['milfoil_presence'] = df['milfoil_presence'].fillna('N')
df
```

```python
df2008 = df.loc[df.year == 2008]
df2008
```

```python
# none, trace, sparse, moderate, dense
milfoil_abundance = ['N', 'T', 'S', 'M', 'D'] # use index for number value

for ma_index, value in enumerate(milfoil_abundance):
  for df_index, row in df2008.iterrows():
    print(df_index) # 0, 1, 2 ... 302, 303
    # print(df2008.index)
    print(df2008.milfoil_presence)

# for ma_index, value in enumerate(milfoil_abundance):
#   for df_index, presence in enumerate(df.milfoil_presence):
#     if presence == value:
#       print(ma_index, df['latitude'][df_index], df['longitude'][df_index])
```

```python
for ma_index, value in enumerate(milfoil_abundance):
  for df_index, row in df2008.iterrows():
    if row['milfoil_presence'] == value:
      # print(row['latitude'], type(row['latitude']), row['longitude'], type(row['longitude']))
      folium.features.CircleMarker(
        [row['latitude'], row['corrected_longitude']],
        radius=ma_index*2,
        color=None, #'green', 
        fill_color='green', 
        fill_opacity=0.5,
    ).add_to(lake_map_2008)
lake_map_2008
```
